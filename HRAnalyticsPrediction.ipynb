```python
X.shape
```

```python
y.shape
```

```python
sns.countplot(y)
```

```python
## 2. skewness have to be reduced
from sklearn.preprocessing import PowerTransformer
```

```python
pt=PowerTransformer(method='yeo-johnson')
```

```python
X=pd.DataFrame(pt.fit_transform(X),columns=X.columns)
```

```python
X
```

```python
### Now checking skewness
X.skew()
```

```python

plt.figure(figsize=(20,25),facecolor='pink')
plotnumber=1

for column in X:
    if plotnumber <= 35:
        ax=plt.subplot(5,7,plotnumber)
        sns.distplot(X[column],color='k')
        plt.xlabel(column,fontsize=14)
    plotnumber=plotnumber+1
plt.show()
```

```python
### 3. Dynamic reduction through PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python

scaler=StandardScaler()
pca=PCA()
```

```python
x_scaled=scaler.fit_transform(X)
```

```python
pca.fit_transform(x_scaled)
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('PCA Component')
plt.ylabel('Varience Covered')
plt.title('PCA')
plt.show()
```

# Considering 23 component will covered the 98% varience in the data point.

```python
pca=PCA(n_components=23)
new_comp=pca.fit_transform(x_scaled)
```

```python
p_c=pd.DataFrame(new_comp,columns=['PC'+ str(i) for i in range(0,23)])
```

```python
p_c
```

```python
p_c.corrwith(y).plot(kind='bar',grid=True,figsize=(15,7),title='Correlation with Target')
```

```python
plt.figure(figsize=(30,20))
sns.heatmap(p_c.corr(),annot=True,annot_kws={'size':10},linewidth=0.6,fmt='0.2f')
plt.show()
```

## PCA reduced the multicolinearity problem. The preprocessing is done now,have to start model buiding

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,roc_auc_score,confusion_matrix,classification_report,plot_roc_curve
```

```python
x_train,x_test,y_train,y_test=train_test_split(p_c,y,test_size=0.25,random_state=446)
```

```python
def model(clf,x_train,x_test,y_train,y_test,train=True):
    if train:
        ypred=clf.predict(x_train)
        print('  Training Result  ')
        print('Accuracy_Score: ',accuracy_score(y_train,ypred)*100)
        
    elif train==False:
        y_pred=clf.predict(x_test)
        print('   Test Result    ')
        print('Accuracy_Score: ',accuracy_score(y_test,y_pred)*100)
        print(confusion_matrix(y_test,y_pred))
        print(classification_report(y_test,y_pred))
        plt.figure(figsize=(15,7))
        sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,annot_kws={'size':10},fmt='0.2f')
        
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
model(lr,x_train,x_test,y_train,y_test,train=True)
```

```python
model(lr,x_train,x_test,y_train,y_test,train=False)
```
